```python
duplicate_rows_data=data[data.duplicated()]
print("number of duplicate rows:",duplicate_rows_data.shape) 
```

```python
data=data.drop_duplicates()
data.shape
```

```python
data.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("no of car by make")
plt.xlabel("no of cars") 
plt.ylabel(" make")
```

```python
data.head(5)
```

```python
data['tot']=data['Engine HP']+data['MSRP']
data.head(5)
```

```python
def get_price_category(MSRP):
    if MSRP<30000:
        return "budget"
    elif 30000<=MSRP<60000:
        return "medium"
    else:
        return "high-end"
data['price_category']=data['MSRP'].apply(get_price_category)
data.head(5)
```

#merging and joining

```python
agg_df=data.groupby('price_category').agg({'MSRP':'mean','Engine HP':'mean'})
agg_df.reset_index(inplace=True)
agg_df.head(5)
```
