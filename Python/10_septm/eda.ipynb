```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv(r"D:\basic\Python\10_septm\data.csv")
data.head(5)
```

```python
print(data.isnull().sum())
```

```python
data=data.dropna()
data.shape
```

```python
duplicate_rows_data=data[data.duplicated()]
print("number of duplicate rows:",duplicate_rows_data.shape) 
```

```python
data=data.drop_duplicates()
data.shape
```

```python
data.count()
```

```python
data.info()
```

```python
data.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("no of car by make")
plt.xlabel("no of cars") 
plt.ylabel(" make")
```

```python
Q1=data['MSRP'].quantile(0.25)
Q3=data['MSRP'].quantile(0.75) 
IQR=Q3-Q1
print(IQR)
sns.boxplot(x=data['MSRP'])
```

```python
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
```
